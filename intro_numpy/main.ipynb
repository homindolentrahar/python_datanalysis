## Boolean arrays

```python
a = np.arange(4)
```

```python
a
```

```python
a[0], a[-1]
```

```python
a[[0, -1]]
```

```python
a[[True, False, False, True]]
```

```python
a
```

```python
a >= 2
```

```python
a[a >= 2]
```

```python
a.mean()
```

```python
a[a > a.mean()]
```

```python
a[~(a > a.mean())]
```

```python
a[(a == 0) | (a == 1)]
```

```python
a[(a <= 2) & (a % 2 == 0)]
```

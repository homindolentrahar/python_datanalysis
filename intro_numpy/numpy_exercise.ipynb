## Boolean arrays

```python
X = np.array([-1, 2, 0, -4, 5, 6, 0, 0, -9, 10])

mask = X <= 0
mask
```

```python
X = np.array([-1, 2, 0, -4, 5, 6, 0, 0, -9, 10])

mask = X > 5
X[mask]
```

```python
X = np.array([-1, 2, 0, -4, 5, 6, 0, 0, -9, 10])

mask = X > X.mean()
X[mask]
```

```python
X = np.array([-1, 2, 0, -4, 5, 6, 0, 0, -9, 10])

mask = (X == 2) | (X == 10)
X[mask]
```

## Logic functions

```python
X = np.array([-1, 2, 0, -4, 5, 6, 0, 0, -9, 10])

X.all()
```

```python
X = np.array([-1, 2, 0, -4, 5, 6, 0, 0, -9, 10])

X.any()
```
